Exercise 1: Revise a previous program as follows: Read and parse the
“From” lines and pull out the addresses from the line. Count the number of messages from each person using a dictionary.

After all the data has been read, print the person with the most commits
by creating a list of (count, email) tuples from the dictionary. Then
sort the list in reverse order and print out the person who has the most
commits

```
Sample Line:
From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008
Enter a file name: mbox-short.txt
cwen@iupui.edu 5
Enter a file name: mbox.txt
```

````python
import re

fname = input("Enter a file name: (HINT:mbox.txt or mbox-short.txt)")
try: 
    fhand = open(fname)
except: 
    print('file cannot be opened:', fname)
    exit()

counts = dict()
for line in fhand:
  line = line.rstrip()
  # Search for lines that Match start with “From:“ and sign @    
  if re.search('^From:.+@', line):
    # word split from line
    words = line.split()
    if words[1] in counts:
      counts[words[1]] += 1 
    else:
      counts[words[1]] = 1

# The person with the most commits by creating a list of tuples from the dictionary counts().    
lst = list()
for key,val in list(counts.items()):
  lst.append((val,key))
lst.sort(reverse=True)
# Display the email that has the most commits
for key,val in lst[:1]:
  print(val,key) 
````

# py4e(pdf): Exercises 2, p.128

Exercise 2: This program counts the distribution of the hour of the day
for each of the messages. You can pull the hour from the “From” line
by finding the time string and then splitting that string into parts using
the colon character. Once you have accumulated the counts for each
hour, print out the counts, one per line, sorted by hour as shown below.

```
python timeofday.py
Enter a file name: mbox-short.txt

04 3
06 1
07 1
09 2
10 3
11 6
14 1
15 2
16 4
17 2
18 1
19 1

```

````python
import re

fname = input("Enter a file name: (HINT: mbox-short.txt)")
try: 
    fhand = open(fname)
except: 
    print('file cannot be opened:', fname)
    exit()

counts = dict()
for line in fhand:
  line = line.rstrip()
  # Search for lines that start with From and a character 
  # followed by a two digit number between 00 and 99 followed by ':'
  hour = re.findall('^From .* ([0-9][0-9]):', line)
  # filter the empty
  if len(hour) > 0:     
    if hour[0] in counts:
      counts[hour[0]] += 1 
    else:
      counts[hour[0]] = 1

#The accumulated the counts for each hour by creating a list of tuples from the dictionary counts().    
lst = list()
for key,val in list(counts.items()):
  lst.append((key,val))
# Sort the list 
lst.sort()
# Display accumulated the counts for each hour
for key,val in lst[:]:
  print(key,val)     
````

# py4e(pdf): Exercise 3, p.129

Exercise 3: Write a program that reads a file and prints the letters
in decreasing order of frequency. Your program should convert all the
input to lower case and only count the letters a-z. Your program should
not count spaces, digits, punctuation, or anything other than the letters
a-z. Find text samples from several different languages and see how
letter frequency varies between languages. Compare your results with
the tables at https://wikipedia.org/wiki/Letter_frequencies.

````python
# download the sample file (romeo full)
!wget https://www.py4e.com/code3/romeo-full.txt
````

````python
import re

fname = input("Enter file name: (HINT:romeo-full.txt)")
try: 
    fhand = open(fname)
except: 
    print('file cannot be opened:', fname)
    exit()

letter_counts = 0
counts = dict()
for line in fhand:
  line = line.rstrip()
  # convert to lowercase 
  line = line.lower()
  # Search for lines that all lower case letters
  word = re.findall('([a-z])', line)
  # unmatch the empty word for each line  
  if len(word) > 0:
    # Count totals letter for each line    
    letter_counts += len(word)        
    for letter in word:    
      # Count number for each letter     
      if letter in counts:
        counts[letter] += 1
      else:
        counts[letter] = 1

#The counts the letter and frequency by creating a list of tuples from the dictionary counts(). 
lst = list()
for key,val in list(counts.items()):
  lst.append((key,val))
# Sort the list
lst.sort()
# Display letter and frequency from input file
for key,val in lst[:]:
  print(key, '{}/{}'.format(val, letter_counts) ,'({:.2%})'.format(val/letter_counts))   
````
